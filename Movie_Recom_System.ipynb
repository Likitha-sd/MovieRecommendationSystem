```python
# Converting string to list
import ast

def convert(text):
    try:
        parsed = ast.literal_eval(text)
        result = []
        for item in parsed:
            if isinstance(item, dict) and 'name' in item:
                result.append(item['name'])
            elif isinstance(item, str):
                result.append(item)
        return result
    except (ValueError, SyntaxError):
        # If parsing fails, just return the original text
        return text
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
def convert_cast(text):
    # If it's already a list, just slice the first 3
    if isinstance(text, list):
        return text[:3]
    
    # If it's a string, parse it safely
    try:
        parsed = ast.literal_eval(text)
        return [item['name'] if isinstance(item, dict) else item for item in parsed[:3]]
    except (ValueError, SyntaxError):
        return []
```

```python
movies = movies.dropna(subset=['cast'])
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
# fetching director name
import ast

def fetch_director(text):
    directors = []
    try:
        crew_list = ast.literal_eval(text)  # safely parse the string
        for i in crew_list:
            # Case 1: dictionary with job info
            if isinstance(i, dict) and i.get('job') == 'Director':
                directors.append(i.get('name'))
            # Case 2: plain string (already a director name)
            elif isinstance(i, str):
                directors.append(i)
    except (ValueError, SyntaxError):
        # If parsing fails, just return empty list
        return []
    return directors
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
# Handling Overview
movies.iloc[0]['overview']
```

```python
movies['overview'] = movies['overview'].fillna("").apply(
    lambda x: x.split() if isinstance(x, str) else x
)

movies.sample(4)
```

```python
def remove_spaces(L):
    if isinstance(L, list):  # case 1: list of names
        return [i.replace(" ", "") for i in L]
    elif isinstance(L, str):  # case 2: single string
        return L.replace(" ", "")
    else:  # case 3: NaN or other types
        return []
```

```python
movies['cast']=movies['cast'].apply(remove_spaces)
movies['crew']=movies['crew'].apply(remove_spaces)
movies['genres']=movies['genres'].apply(remove_spaces)
movies['belongs_to_collection']=movies['belongs_to_collection'].apply(remove_spaces)
```

```python
movies.head()
```

```python
# concatenate
def ensure_list(x):
    if isinstance(x, list):
        return x
    elif isinstance(x, str):
        return [x]
    else:
        return []

movies['overview'] = movies['overview'].apply(ensure_list)
movies['belongs_to_collection'] = movies['belongs_to_collection'].apply(ensure_list)
movies['cast'] = movies['cast'].apply(ensure_list)
movies['genres']=movies['genres'].apply(ensure_list)
movies['tags'] = movies['overview'] + movies['belongs_to_collection'] + movies['cast']+movies['genres']
```

```python
# drop few unwanted columns
new_df=movies[['id', 'title', 'tags']]
new_df.head()
```

```python
# converting list to string
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
# stemming
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    T = []
    # if text is already a list of tokens
    if isinstance(text, list):
        for i in text:
            T.append(ps.stem(i))
    # if text is a string
    elif isinstance(text, str):
        for i in text.split():
            T.append(ps.stem(i))
    return " ".join(T)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
# Recommendation Function
def recommend(movie):
    if movie not in new_df['title'].values:
        print("movie not found")
        return
    index=new_df[new_df['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend(input('Enter a movie to get recommendation:'))
```

```python
import streamlit as st
import pickle as pk

movies = pd.read_pickle(r"D:/MovieRecomSys/movie_list (1).pkl")
similarity = pickle.load(open(r"D:/MovieRecomSys\similarity (1).pkl", "rb"))




def recommend(movie):
    try:
        index = movies[movies['title'].str.lower() == movie.lower()].index[0]
    except IndexError:
        return ['Movie not found'] 
    
    distances = sorted(enumerate(similarity[index]), key=lambda x: x[1], reverse=True)
    return [movies.iloc[i[0]].title for i in distances[1:6]]

st.title('Movie Recommendation System')
movie_list = movies['title'].values
selected_movie = st.selectbox('Select a Movie', movie_list)

if st.button('Recommend'):
    recommendations = recommend(selected_movie)
    for movie in recommendations:
        st.write(movie)
```
